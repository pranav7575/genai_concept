Stemming 

```python
from nltk.stem import PorterStemmer, RegexpStemmer,SnowballStemmer
stemmer = PorterStemmer()
```

```python
words=["running", "jumps", "easily", "fairly",'history','studies']
```

```python
for word in words:
    print(word+"--->"+ stemmer.stem(word))
```

```python
# Using RegexpStemmer
rex_stemmer = RegexpStemmer('ing$|s$|e$|able$|ly$|ity$|ed$')
```

```python
for word in words:
    print(word+" --> "+rex_stemmer.stem(word))
```

```python
SnowballStemmer_var=SnowballStemmer("english")
for word in words:
    print(word+" --> "+SnowballStemmer_var.stem(word))
```
